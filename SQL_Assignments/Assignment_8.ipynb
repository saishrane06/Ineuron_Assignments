## Q-15. Write an SQL query to fetch intersecting records of two tables.

select w.id,w.First_Name,wc.id,wc.First_Name from Worker w inner join Worker_Clone wc on w.id=wc.id;

## Q-16. Write an SQL query to show records from one table that another table does not have.

select w.First_Name,w.Last_Name,wc.First_Name,wc.Last_Name from Worker w left join Worker_Clone wc on w.id=wc.id;
