## Q-19. Write an SQL query to determine the 5th highest salary without using TOP or limit method.

select distinct Salary from Worker w1 where 5 = (select count(distinct Salary) from Worker w2 where w2.salary >= w1.salary);

## Q-20. Write an SQL query to fetch the list of employees with the same salary.

select distinct w.First_Name,w.Salary from Worker w,Worker w2 where w.Salary=w2.Salary and w.First_Name <> w2.First_Name;
